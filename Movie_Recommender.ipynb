# DATA PREPROCESSING

```python
def convert(obj):
    L = []
    for i in (obj):
        L.append(i["name"])
    return L
        
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
movies["genres"]=movies["genres"].apply(convert)
```

```python
movies
```

```python
movies['keywords'][0]
```

```python
movies["keywords"]=movies["keywords"].apply(convert)
```

```python
movies["cast"][0]
```

```python
def convert3(obj):
    L = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i["name"])
            counter+=1
        else:
            break
    return L
```

```python
movies["cast"]=movies["cast"].apply(convert3)
```

```python
movies.head()
```

```python
movies["crew"][0]
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            L.append(i["name"])
            break      
    return L
```

```python
movies["crew"]=movies["crew"].apply(fetch_director)
```

```python
movies.head()
```

```python
movies["overview"][0]
```

```python
movies["overview"] = movies["overview"].apply( lambda x:x.split())
```

```python
movies.head()
```

```python
# we need to remove space between two words from genres, keywords , cast and crew 
```

```python
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
# now create tag column which contain keyword,genres,overview,cast and crew with the help of concatination
```

```python
movies["tags"]= movies["keywords"] + movies["genres"] + movies["overview"] + movies["cast"] + movies["crew"]
```

```python
#alcohol','alcoholic','alcoholism', all this words has same meaning so we perform steming to consider it as a single word 
# like [" loving","loved","love"]=["love","love","love"]
```

to perform steming we import nltk lab

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df["tags"] = new_df["tags"].apply(stem)
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer (max_features= 5000 , stop_words="english")
```

```python
vectors=cv.fit_transform(new_df["tags"]).toarray()
```
